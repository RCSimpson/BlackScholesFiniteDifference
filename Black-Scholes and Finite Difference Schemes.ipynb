## Black-Scholes Partial Differential Equation and Finite Difference Schemes

We can use finite difference schemes to solve for the Black-Scholes Partial Differntial equation. While the Black Scholes equation does have an analytical solution for the European Call option this isn't necessarily the case for all variants. We can use finite difference Schemes to solve the Black Scholes Equation in the following way.

We recall that the Black-Scholes Partial Differential Equation is as follows: 

$$
V_{t} + \frac{1}{2} \sigma^{2} S^{2} V_{SS} - rV + rS V_{S} = 0
$$

These derivatives are commonly known to be the Greeks, parameters describing characteristics of the options, but if we treat them as variables we can derive the partial differential equation above. 

### Explicit Finite Difference Scheme

We can approximate local deratives at points in our time-price space.

$$
V_{t} \approx \frac{v^{n+1}_{m} - v^{n}_{m} }{k}
$$


And in the stock price $S$ the finite differences are as follows:

Central:

$$
V_{S} \approx \frac{v^{n}_{m+1} - v^{n}_{m-1} }{2h}
$$


Forward:

$$
V_{SS} \approx \frac{v^{n}_{m+1} - 2v^{n}_{m}  + v^{n}_{m-1} }{h^2}
$$

Using these approximations we can substitute them into the Black-SCholes equation above and after some difficult algebra we find that: 
    
    
Let $\alpha = \sigma^2 k $ and $\beta = r k $.


\begin{align*}
    & v^{n+1}_{m} = (1 - \alpha m^2  - \beta )v^{n}_{m} + \frac{1}{2}\left( \alpha m^2 + \beta m \right)v^{n}_{m+1} + \frac{1}{2}\left( \alpha m^2 - \beta m \right)v^{n}_{m-1}
\end{align*}

We can then use this formula to solve for the Black-Scholes solution.
